```python
INPUT_FILE ="C:/Users/alvar/OneDrive/Personal/Repositorio/Data Analytics/Kaggle/titanic/train.csv"
OUTPUT_FILE_PANDAS_PROFILING ="pandas_profiling.html"
```

## [Pandas Profiling](#https://pandas-profiling.github.io/pandas-profiling/docs/master/rtd/pages/introduction.html)

```python
pip install pandas-profiling
```

```python
import pandas as pd
from pandas_profiling import ProfileReport

#EDA using pandas-profiling
profile = ProfileReport( pd.read_csv( INPUT_FILE ), explorative= True )

# Saving results to a HTML file
profile.to_file( OUTPUT_FILE_PANDAS_PROFILING )
```

## [Sweetviz](#https://github.com/fbdesignpro/sweetviz)

```python
pip install sweetviz
```

```python
import pandas as pd
import sweetviz as sv

#EDA using Sweetviz
sweet_report = sv.analyze( pd.read_csv( INPUT_FILE ) )

# Saving results to HTML file
sweet_report.show_html( )
```

## [Autoviz](#https://github.com/AutoViML/AutoViz/blob/master/README.md)

```python
pip install autoviz
```

```python
import pandas as pd
from autoviz.AutoViz_Class import AutoViz_Class

print( INPUT_FILE )
print( OUTPUT_FILE_PANDAS_PROFILING )

# EDA using AutoViz
autoviz = AutoViz_Class().AutoViz( INPUT_FILE )
```
